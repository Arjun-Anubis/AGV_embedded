```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
df = pd.read_csv("out.csv")
```

```python
df["x"] = 4095 * np.cos( np.pi/2 * df["angle"]/255 )
df["y"] = 4095 * np.sin( np.pi/2 * df["angle"]/255 )
```

```python
df["err_x"] = (df["x"] - df[" x"]) / (df[" x"])
df["err_y"] = (df["y"] - df[" y"]) / (df[" y"])
```

```python
df.plot.scatter( x = "x", y = " x" )
```

```python
df.plot.scatter( x = "y", y = " y" )
```
